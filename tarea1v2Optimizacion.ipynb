```python
op = ["/","+", "-", "*"]
num = ["1","2", "3", "4", "5","6", "7", "8" , "9" ]
```

```python
# Valor a buscar
busc=77
```

#Resultados 1

En esta epata se prueba el programa principal, o base para enceontrar el valor buscado, tambien en este punto es calcularemos empiricamente la cantidad total de valores y los minimos y maximos.

```python
alt1=""
contador=0
solucion=[]

minimo =1000
maximo=0

for n1 in (list(set(num)-set(alt1))):
  alt1 = n1
  for o1 in (list(set(op)-set(alt1))):
    alt2 = o1+alt1
    for n2 in (list(set(num)-set(alt2))):
      alt3 = n2+alt2
      for o2 in (list(set(op)-set(alt3))):   
        alt4 = o2+alt3 
        for n3 in (list(set(num)-set(alt4))): 
          alt5 = n3+alt4           
          for o3 in (list(set(op)-set(alt5))):    
            alt6 = o3+alt5        
            for n4 in (list(set(num)-set(alt6))): 
              alt7 = n4+alt6
              for o4 in (list(set(op)-set(alt7))):   
                alt8 = o4+alt7
                for n5 in (list(set(num)-set(alt8))): 
                  alt9 = n5+alt8
                  valor= eval(alt9)
                  if valor == busc:
                    solucion.append(alt9)
                  contador+=1
                  if valor < minimo:
                    minimo=valor
                  if maximo < valor:
                    maximo=valor              

print("Cantidad  de calculos: ", contador)   
print("Valor minimo calculable: ",   minimo)      
print("Valor maximo calculable: ", maximo)     
print("")
print("Soluciones para : ", busc)
print("")

for s in solucion:
  print(s)        


```

# Resultados 2
##Revision de espectro de numeros posibles y su factibilidad de obtencion.

Se aprecia del grafico que dentro del rango factible minimo y maximo, se pueden enceontrar todos los numeros enteros, siendo los valores interiores, los que tiene mas combinaciones posibles, esto ya que los extremos, necesitas de las operaciones mas potentes, equivalentes a multiplicacion de los numeros de mayor magnitud


```python
# Generamos una fucnion que nos devolvera una lista de alternatias posibles

def combinacion(busc, op, num):

  alt1=""
  contador=0
  solucion=[]

  for n1 in (list(set(num)-set(alt1))):
    alt1 = n1
    for o1 in (list(set(op)-set(alt1))):
      alt2 = o1+alt1
      for n2 in (list(set(num)-set(alt2))):
        alt3 = n2+alt2
        for o2 in (list(set(op)-set(alt3))):   
          alt4 = o2+alt3 
          for n3 in (list(set(num)-set(alt4))): 
            alt5 = n3+alt4           
            for o3 in (list(set(op)-set(alt5))):    
              alt6 = o3+alt5        
              for n4 in (list(set(num)-set(alt6))): 
                alt7 = n4+alt6
                for o4 in (list(set(op)-set(alt7))):   
                  alt8 = o4+alt7
                  for n5 in (list(set(num)-set(alt8))): 
                    alt9 = n5+alt8
                    valor= eval(alt9)
                    if valor == busc:
                      solucion.append(alt9)
                      contador+=1
        
  return busc, contador
```

```python
lisComp =[]

for n in range(-73,81):
  r=combinacion(n, op, num)
  lisComp.append(r)

```

```python
# Lista de valores como tuplas (vslor buscsdo, csntidad de combinaciones posibles)
print(lisComp)
```

```python
import matplotlib.pyplot as plt

x, y = zip(*lisComp)
plt.scatter(x, y)

plt.xlabel("Valores a Encontrar")
plt.ylabel("Cantidad de Alternativas")

plt.title("Valores y Alternativas Encontradas")

plt.show()
```
